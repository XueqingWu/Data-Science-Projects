**Identify and Handel Missing Value**

```python
train.shape
```

```python
Missing_Val_Col=[]
for i in train.columns:
    Null_Sum=train[i].isnull().sum()
    Null_Per=Null_Sum/40000
    if Null_Sum!=0:
        print(i, Null_Sum, Null_Per)
        if Null_Per >0.5:
            Missing_Val_Col.append(i)
            
Missing_Val_Col
```

```python
train.drop(Missing_Val_Col, axis=1, inplace=True)
```

```python
train.shape
```

```python
numerical = [var for var in train.columns if train[var].dtype!='O']
print(numerical)
print('There are {} numerical variables'.format(len(numerical)))
```

```python
categorical = [var for var in train.columns if train[var].dtype=='O']
print(categorical)
print('There are {} categorical variables'.format(len(categorical)))
```

```python
# fill the missing value with median
for col in numerical:
    if train[col].isnull().sum()>0:
        print(col)
        train[col].replace(np.nan, train[col].median(), inplace=True)
```

```python
Cat_Missing=[]
for col in categorical:
    if train[col].isnull().sum()>0:
        Cat_Missing.append(col)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix

model=LogisticRegression()
model.fit(train_x,train_y)
prediction=model.predict(test_x)
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction,test_y))
```

```python
prediction
```

```python
array = confusion_matrix(test_y,prediction)
df_cm = pd.DataFrame(array, index = test['TARGET'].unique(),
                  columns = test['TARGET'].unique())
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True)
```

```python
from sklearn.metrics import classification_report
print("Logistic Regression",classification_report(test_y, prediction, target_names=None))
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction,test_y)) 
```

**K-Nearest Neighbor**

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
model.fit(train_x,train_y)
prediction=model.predict(test_x)
print('The accuracy of the K-Nearest Neighbours is',metrics.accuracy_score(prediction,test_y))
```

```python
array = confusion_matrix(test_y,prediction)
df_cm = pd.DataFrame(array, index = test['TARGET'].unique(),
                  columns = test['TARGET'].unique())
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True)
```

```python
print("K Nearest Neighbor",classification_report(test_y, prediction, target_names=None))
print('The accuracy of the K Nearest Neighbor is',metrics.accuracy_score(prediction,test_y)) 
```

**Decision Tree**

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(train_x, train_y)
#tree.plot_tree(clf)   ##show tree
prediction=clf.predict(test_x)
print('The accuracy of Decision tree is',metrics.accuracy_score(prediction,test_y))
```

```python
array = confusion_matrix(test_y,prediction)
df_cm = pd.DataFrame(array, index = ['Test', 'Prediction'],
                  columns = [0, 1])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True)
```

```python
print("Decision Tree",classification_report(test_y, prediction, target_names=None))
print('The accuracy of the Decision Tree is',metrics.accuracy_score(prediction,test_y)) 
```
